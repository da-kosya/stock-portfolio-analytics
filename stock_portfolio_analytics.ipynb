```python
import yfinance as yf
import pandas as pd 

# getting the 1 year historical data of Apple
apple = yf.Ticker('AAPL')
data = apple.history(start='2024-01-01', end='2025-01-01')

print(data.head())
print(data.dtypes)
print(data.describe())
```

```python
# single best day for AAPL last year
# daily return: how much did my money grow or shrink today, as a percentage? 
# daily return = (today's price - yesterday's price) / yesterday's price
data['daily_return'] = data['Close'].pct_change()
print(data['daily_return'].max())
print(data['daily_return'].min())
print(data['daily_return'].mean())

print(data[['Close', 'daily_return']].head(10))
```
